```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read movies.txt
movies = open('../src/data/movies.txt').read()
```

```python
wordcloud = WordCloud(max_words=100, background_color="white").generate(movies)
```

```python
plt.imshow(wordcloud)
plt.axis("off")
plt.show()

# Save the word cloud image
wordcloud.to_file('../output/figs/movies_wordcloud.png')
```
