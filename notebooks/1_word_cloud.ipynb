```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read movies.txt
movies = open('../src/data/movies.txt').read()
```

```python
from wordcloud import STOPWORDS

wordcloud = WordCloud(
    max_words=100, 
    background_color="white",
    width=800,
    height=600,
    stopwords=STOPWORDS.union({"mr", "mrs", "ms", "mss", "de"})
    ).generate(movies)
```

```python
plt.imshow(wordcloud)
plt.axis("off")
plt.show()

# Save the word cloud image
wordcloud.to_file('../output/figs/movies_wordcloud.png')
```
